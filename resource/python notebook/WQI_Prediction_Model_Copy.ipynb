```python
#import stuff

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv as csv
import array
import math

#must need for preprocessing and evaluation
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from matplotlib import pyplot
from scipy import stats

#libraries for the ML model
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor

#libraries for exporting
from joblib import dump, load

#testing best attributes
from sklearn.model_selection import GridSearchCV


stdScaler = StandardScaler()

water_data = pd.read_csv("water_data_1.0.csv")
water_data = water_data.sort_values(by=['STATION CODE','year'])
```

```python
def getFeature(stationCode, year, feature):
    
    try:
        row = water_data.loc[(water_data['STATION CODE'] == str(stationCode))  & (water_data['year'] == year)]   
        return row.iloc[0][feature]
    except:
        return None
```

```python
water_data_time = water_data
water_data_time = water_data_time.drop(columns = ['LOCATIONS','STATE','Temp'])
water_data_time.dropna()
water_data_time = water_data_time.sort_values(by=['STATION CODE','year'])

water_data_time.to_csv(r'C:\Users\USER\Desktop\water_time.csv', index = False, header=True)
```

```python
#adding time

water_data_time = water_data
water_data_time = water_data_time.drop(columns = ['LOCATIONS','STATE','Temp'])
water_data_time.dropna()
water_data_time = water_data_time.sort_values(by=['STATION CODE','year'])

#stationCode_and_year = zip(water_data['STATION CODE'], water_data['year'])

water_data_time['WQI t-1'] = [getFeature(code,year-1,'WQI') for code, year in zip(water_data['STATION CODE'], water_data['year'])]
water_data_time['WQI t-2'] = [getFeature(code,year-2,'WQI') for code, year in zip(water_data['STATION CODE'], water_data['year'])]
water_data_time['WQI t-3'] = [getFeature(code,year-3,'WQI') for code, year in zip(water_data['STATION CODE'], water_data['year'])]
water_data_time['D.O. t-1'] = [getFeature(code,year-1,'D.O.') for code, year in zip(water_data['STATION CODE'], water_data['year'])]
water_data_time['PH t-1'] = [getFeature(code,year-1,'PH') for code, year in zip(water_data['STATION CODE'], water_data['year'])]
water_data_time['CONDUCTIVITY t-1'] = [getFeature(code,year-1,'CONDUCTIVITY') for code, year in zip(water_data['STATION CODE'], water_data['year'])]
water_data_time['B.O.D. t-1'] = [getFeature(code,year-1,'B.O.D.') for code, year in zip(water_data['STATION CODE'], water_data['year'])]
water_data_time['NITRATE t-1'] = [getFeature(code,year-1,'NITRATE') for code, year in zip(water_data['STATION CODE'], water_data['year'])]
water_data_time['FECAL COLIFORM t-1'] = [getFeature(code,year-1,'FECAL COLIFORM') for code, year in zip(water_data['STATION CODE'], water_data['year'])]
water_data_time['TOTAL COLIFORM t-1'] = [getFeature(code,year-1,'TOTAL COLIFORM') for code, year in zip(water_data['STATION CODE'], water_data['year'])]

water_data_time
```

```python
#preprocessing

temp = water_data_time.dropna()

temp = temp.drop(columns = ['STATION CODE','year','D.O.','PH','CONDUCTIVITY','B.O.D.','NITRATE','FECAL COLIFORM','TOTAL COLIFORM'])
temp = temp[(np.abs(stats.zscore(temp)) < 3).all(axis=1)]
temp.to_csv(r'C:\Users\USER\Desktop\water_time.csv', index = False, header=True)

X_time = temp.drop(columns = ['Potability','WQI'])
#X_time = X_time.dropna()
X_time = pd.DataFrame(stdScaler.fit_transform(X_time), columns = X_time.columns)

yWQI_time = temp['WQI']
yPot_time = temp['Potability']

#X_time['WQI'] = yWQI_time.tolist()

X_time.to_csv(r'C:\Users\USER\Desktop\feature_time.csv', index = False, header=True)
```

```python
#rbf support vector regression
clf = SVR(kernel = 'rbf', C = 4884, gamma = 0.0001, max_iter = -1, tol = 1e-5)
count = 1

total_rmse = 0
total_mae = 0
total_r2 = 0

for i in range(count):

    X_train, X_test, y_train, y_test = train_test_split(X_time, yWQI_time, test_size = 0.2)

    clf.fit(X_train, y_train)

    predictions = clf.predict(X_test)

    total_rmse += math.sqrt(mean_squared_error(y_test, predictions))
    total_mae += mean_absolute_error(y_test, predictions)
    total_r2 += r2_score(y_test, predictions)

print("root mean squared error: " + str(total_rmse/count))
print("mean absolute error: " + str(total_mae/count))
print("R squared: " + str(total_r2/count))  
```

```python
#export rbf support vector regression
dump(clf, 'WQIModelv3.0.pkl')
```

```python
#MLP Regression
clf = MLPRegressor(activation = 'logistic', hidden_layer_sizes=(7,5), max_iter = 10000)

X_train, X_test, y_train, y_test = train_test_split(X_time, yWQI_time, test_size = 0.2)

clf.fit(X_train, y_train)

predictions = clf.predict(X_test)

MSE = mean_squared_error(y_test, predictions)
MAE = mean_absolute_error(y_test, predictions)
r_squared = r2_score(y_test, predictions)

print("root mean squared error: " + str(math.sqrt(MSE)))
print("mean absolute error: " + str(MAE))
print("R squared: " + str(r_squared)) 
```
