```python
#import all packages and the files

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedShuffleSplit

stdScaler = StandardScaler()
rbScaler = RobustScaler()
mmScaler = MinMaxScaler()

water_data = pd.read_csv('water_potability.csv')
```

```python
#pre processing.

#decides what to do with the null values
#water_data_preprocessed = water_data.dropna()
#water_data_preprocessed = water_data
water_data_preprocessed = water_data.fillna(water_data.mean())
#water_data_preprocessed = water_data_preprocessed.fillna(water_data.median())
#water_data_preprocessed = pd.DataFrame(scaler.fit_transform(water_data_preprocessed), columns = water_data_preprocessed.columns)

#applies scaling
X = water_data_preprocessed.drop(columns=['Potability'])
X = pd.DataFrame(rbScaler.fit_transform(X), columns = X.columns)
#X = pd.DataFrame(mmScaler.fit_transform(X), columns = X.columns)
#X = pd.DataFrame(stdScaler.fit_transform(X), columns = X.columns)

y = water_data_preprocessed['Potability']

X
```

```python
#the actual machine learning. has a for loop to find the average accuracy

total_score = 0

clf = SVC(kernel="rbf", C = 1.03, gamma = 0.25)
#C = 1.03, gamma = 0.25
#clf = SVC(kernel="linear", C = 1.03, gamma = 0.25)

for i in range(1):

        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
       
        clf.fit(X_train, y_train)

        predictions = clf.predict(X_test)

        score = accuracy_score(y_test, predictions)
        total_score += score
        
print(total_score/(i+1))
```

```python
#!!!!DONT RUN UNLESS YOU KNOW WHAT YOU ARE DOING. IT WILL TAKE FOREVER TO COMPUTE

C_range = np.linspace(1, 2, 5)
C_range = C_range[C_range!=0]
gamma_range = np.linspace(0, 1, 5)
gamma_range = gamma_range[gamma_range!=0]
param_grid = dict(gamma = gamma_range, C = C_range)

grid = GridSearchCV(SVC(), param_grid=param_grid, cv=5)
grid.fit(X,y)

print(
    "The best parameters are %s with a score of %0.3f"
    % (grid.best_params_, grid.best_score_)
)
```
